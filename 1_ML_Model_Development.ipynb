The R-squared ($R^2$) score, also known as the coefficient of determination, is a statistical measure used in regression analysis to assess how well a regression model fits the observed data. It represents the proportion of the variance in the dependent variable that is predictable from the independent variables.

$R^2 = 1 - \frac{Sum~of~Squared~Residuals}{Total~Sum~of~Squares} = 1 - \frac{SS_{res}}{SS_{tot}}$

*   **$SS_{res}$ (Sum of Squared Residuals):** The sum of the squared differences between the actual values and the predicted values.
*   **$SS_{tot}$ (Total Sum of Squares):** The sum of the squared differences between the actual values and the mean of the actual values.

**Interpretation of R2:**

*   **$R^2$ values range from 0 to 1.**
*   **$R^2 = 0$:** The model explains none of the variability in the dependent variable.
*   **$R^2 = 1$:** The model explains all of the variability in the dependent variable (a perfect fit).
*   **Values between 0 and 1:** Indicate the proportion of the variance explained by the model. For example, an R2 of 0.75 means that 75% of the variance in the dependent variable is explained by the model.

It's important to note that a high R2 doesn't necessarily mean the model is good or that the independent variables are the cause of the changes in the dependent variable. It's just a measure of how well the model fits the existing data.

Import the necessary libraries needed:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

#### Step 1: Load and visualise the dataset (only once for today)

```python
#df = pd.read_csv('toy_dataset.csv')
df = pd.read_csv("https://raw.githubusercontent.com/hamidmukhtar/datasets/refs/heads/main/toy_dataset.csv")
print("Data shape:", df.shape)
display(df.sample(5)) # Show 5 sample objects

# Split the dataset into features and target variable
X = df[['feature1', 'feature2']]
y = df['Class']

# Plot the data
plt.scatter(X['feature1'], X['feature2'], c=y, cmap='bwr', alpha=0.7)
plt.title('Example toy dataset with two classes based on two features')
plt.xlabel('feature 1'), plt.ylabel('feature 2')
plt.show()
```

No line can classify the objects in two classes with 100% accuracy.
Why?

### Step 2: Split the data into train and test sets

```python
rand_state=50 # You can play with the random state later

# Split the dataset into training and testing sets (70% for training, 30% for training)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=rand_state)
```

### Step 3: Model Development (training and testing)
* [Logistic Regression](https://scikit-learn.org/1.5/modules/linear_model.html#logistic-regression): A binary classifier
* Principle: Assigns a **probability** value to each object.
* If the $probability > 0.5$, it belongs to one class, otherwise to the other class.


1. Logistic Regression with original features
* Choose the Logistic Regression Classifier
* Develop a model using _random_ training points
* The model should create a line of separation between the two classes


```python
# Create a logistic regression model
model = LogisticRegression()

# Train the model with original features
model.fit(X_train, y_train)

# Test model on the test data
y_pred = model.predict(X_test)

# Step 4. Model Evaluation. Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Plot the classification results
plt.scatter(X_test['feature1'], X_test['feature2'], c=y_test, cmap='bwr', alpha=0.7)

# Plot the decision boundary
coef = model.coef_[0]
intercept = model.intercept_
x_values = np.linspace(X_test['feature1'].min(), X_test['feature1'].max(), 100)
y_values = -(coef[0] * x_values + intercept) / coef[1]
plt.plot(x_values, y_values, color='black')

plt.xlabel('feature 1')
plt.ylabel('feature 2')
plt.title('Classification Results with Decision Boundary: Test set')
plt.show()

# Show the instances that have not been classified correctly.
test_data = X_test.copy()
test_data['TrueClass'] = y_test
test_data['PredClass'] = y_pred

missed = test_data[test_data['PredClass'] != test_data['TrueClass']]
missed
```

### Feature Engineering
2. Logistic Regression with Polynomial Features
* Transform the existing features into new ones using a mathematical approach
* The new features may be in higher dimensions leading to better separation
* We will use [Polynomial Features](https://scikit-learn.org/1.5/modules/generated/sklearn.preprocessing.PolynomialFeatures.html) in this example
* We start with degree 2 and increase the degree until we get good results

image.png

```python
from sklearn.preprocessing import PolynomialFeatures
from mlxtend.plotting import plot_decision_regions

# Load your dataset... We already did it above (from df -> X and y)
# Split your data... We already did it above (X, y -> X_train, X_test, y_train, y_test)

# Transform the features using PolynomialFeatures
poly = PolynomialFeatures(degree=4)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

# Create a logistic regression model
model = LogisticRegression()

# Train the model
model.fit(X_train_poly, y_train)

# Make predictions
y_pred = model.predict(X_test_poly)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Determine the objects that were not correctly classified
test_data = X_test.copy()
test_data['TrueClass'] = y_test
test_data['PredClass'] = y_pred
missed = test_data[test_data['PredClass'] != test_data['TrueClass']]

# Plot the classification results
plt.scatter(X_test.iloc[:, 0], X_test.iloc[:, 1], c=y_pred, cmap='bwr', alpha=0.3)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Classification Results with Polynomial Features')
plt.scatter(missed['feature1'], missed['feature2'], c=missed['TrueClass'], cmap="bwr", alpha=1, label="Missed")
#plt.legend()
plt.show()

display(missed)
```

## Underfitting vs Overfitting
* If your model is very simple, it is underfit and cannot classify the data very well.
* If you choose a higher degree, than required for the given complexity of data, your model overfits: it **memorizes** the training set and produces bad results on the test set.
* We require a balance between underfit and overfit.

### Let's move to our next classifier
3. Support Vector Machines [SVM](https://scikit-learn.org/1.5/api/sklearn.svm.html#module-sklearn.svm)
* An algorithm that classifies data by finding an optimal line or hyperplane that maximizes the distance between each class in an N-dimensional space.

```python

from sklearn.svm import SVC

# Load and Split the dataset into training and testing sets

# Create an SVM model
model = SVC(C=1, kernel='rbf')  # You can choose different kernels like 'linear', 'rbf', 'poly', 'sigmoid', etc.

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Plot the classification results
plt.scatter(X_test['feature1'], X_test['feature2'], c=y_test, cmap='bwr', alpha=0.5)

# Plot the decision boundary (This can be more complex for non-linear kernels)
# For linear kernel, we can still visualize the boundary approximately

plot_decision_regions(np.array(X_test), np.array(y_test), clf=model, legend=2)
plt.title('Classification Results with SVM: Test set')
plt.show()

# Show the instances that have not been classified correctly.
test_data = X_test.copy()
test_data['TrueClass'] = y_test
test_data['PredClass'] = y_pred

missed = test_data[test_data['PredClass'] != test_data['TrueClass']]
display(missed)
```

# Neural Networks

Neural Networks grew in popularity exponentially in the years after the 2012 computer vision breakthrough
* computational capabilities now sufficient for training very deep models
* non-linear activation functions are used

Neural networks
* It's an architecture of layers and activation functions that is trained
* Training: A mathematical, iterative optimization process

## A Basic Neural Model

image.png

We see a yellow input layer, a red output layer and just one blue hidden layer in between.
* The individual components are **neurons** (hold information)
* The lines tranfer the information from neuron to neuron.
* **Input layer**: capable of accepting input to the model that is in some kind of shape
* **Hidden layer(s)** attempt to capture patterns hidden within the dataset as a whole through training. Builds a mathematical model
* **Output layer**: generates the final prediction. May contain a large number of neurons.

### A Neuron

image.png

* Each individual neuron holds a vector with input _**x**_.
* _**x**_ is multiplied with a weights vector _**w**_.
* A bias value _**b**_ is added to cater for the noise/offset.
* _**w**_ represents the patterns within the dataset *learnt* by each individual neuron.
* _**w**_ is updated over several iterations (a few to thousands).
 **Note:** This is an extension of the equation of line to a vector.

## Activation Functions
image.png

$ neuron~output (x) = activation (wx + b) $

The activation function can be _linear_, but its magic is obvious when it is _non-linear_.

### Rectified Linear Unit (ReLU) Activation Function
image.png

When the input $x<0$, the output is $0$. Otherwise, it is $x$.

### Working with Non-Linear Data

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200, random_state=rand_state)

# Split the dataset into training and testing sets
# Caution: This is re-assigning the same variables to new data!!!
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=rand_state)

# Generate scatter plot for training data
plt.scatter(X_train[:,0], X_train[:,1])
plt.title('Non-linear data')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

### Required imports for Deep Learning models using the Keras/Tensorflow libraries

```python
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from mlxtend.plotting import plot_decision_regions
```

### Let's create a very basic neural network: A Perceptron
It has only an input and an output layer (connects the input directly to the output)

No hidden layer $→$ no learning

```python
#The model requires the dimensions of the input features
input_shape = (X_train.ndim,)
print(f'Feature shape: {input_shape}')

# Create the model
model = Sequential()
model.add(Dense(1, input_shape=input_shape, activation='sigmoid')) #input layer directly connected to the output

# Configure the model and start training
model.compile(loss='squared_hinge', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=1, validation_split=0.2)

# Test the model after training
test_results = model.evaluate(X_test, y_test, verbose=1)
print(f'Test results - Loss: {test_results[0]} - Accuracy: {test_results[1]*100}%')

# Plot decision boundary
plot_decision_regions(X_test, y_test, clf=model, legend=2)
plt.show()
```

The output is simply a **chance** (50/50)

### Let's add a hidden layer
A neural network with:
* An **input** layer taking the input shape of images as input and with 50 neurons (units) to process the input attributes
* One **hidden** layer with `Linear` activation function (compare it with non-linear activation function later)
* One **output** layer with `Sigmoid` activation function (for binary classification)

A lot can be configured in this network (kernel_initializer, loss function, optimizer, metrics, epochs, batch_size,...)

```python
#The model requires the dimensions of the input features
input_shape = (X_train.ndim,)
print(f'Feature shape: {input_shape}')

# Create the model
model = Sequential()
model.add(Dense(50, input_shape=input_shape, activation='linear', kernel_initializer='he_uniform')) #hidden layer
model.add(Dense(1, activation='sigmoid')) # output layer

# Configure the model and start training
model.compile(loss='squared_hinge', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=1, validation_split=0.2)

# Test the model after training
test_results = model.evaluate(X_test, y_test, verbose=1)
print(f'Test results - Loss: {test_results[0]} - Accuracy: {test_results[1]*100}%')

# Plot decision boundary
plot_decision_regions(X_test, y_test, clf=model, legend=2)
plt.show()
```

No linear function can get better results. Let's

### Build a complex neural network
* Add more hidden layers (3 layers)
* Add **relu** activation function (compared to linear activation above) in the hidden layers

image.png

```python
model = Sequential()
model.add(Dense(200, input_shape=input_shape, activation='relu', kernel_initializer='he_uniform')) #input layer
model.add(Dense(150, activation='relu', kernel_initializer='he_uniform')) #hidden layer 1
model.add(Dense(100, activation='relu', kernel_initializer='he_uniform')) #hidden layer 2
model.add(Dense(50, activation='relu', kernel_initializer='he_uniform')) #hidden layer 3
model.add(Dense(1, activation='sigmoid')) #change activation to tanh?

model.compile(loss='squared_hinge', optimizer='adam', metrics=['accuracy'])   # change the optimizer
model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=1, validation_split=0.2)

# Test the model after training
loss, accuracy = model.evaluate(X_test, y_test, verbose=1)
print(f'Test results - Loss: {loss} - Accuracy: {accuracy*100}%')

# Plot decision boundary
plot_decision_regions(X_test, y_test, clf=model, legend=2)
plt.show()
```
